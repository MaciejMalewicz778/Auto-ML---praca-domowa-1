```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

bayes_lr = pd.read_csv('bayes_results_lr.csv')
bayes_tree = pd.read_csv('bayes_results_tree.csv')
bayes_forest = pd.read_csv('bayes_results_forest.csv')
```

```python
plt.figure(figsize=(17, 3))
for i in range(1, 5):
    plt.subplot(1, 4, i)
    bayes_lr[f'wynik_zbior_{i}'].plot(title=f'Zbiór numer {i}')

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.suptitle("Regresja logistyczna", bbox=props, verticalalignment='top', x=0.51, y=0.965)

plt.savefig('plots/convergance_lr', bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(17, 3))
for i in range(1, 5):
    plt.subplot(1, 4, i)
    bayes_tree[f'wynik_zbior_{i}'].plot(title=f'Zbiór numer {i}')

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.suptitle("Drzewo losowe", bbox=props, verticalalignment='top', x=0.51, y=0.965)

plt.savefig('plots/convergance_tree', bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(17, 3))
for i in range(1, 5):
    plt.subplot(1, 4, i)
    bayes_forest[f'wynik_zbior_{i}'].plot(title=f'Zbiór numer {i}')

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.suptitle("Las losowy", bbox=props, verticalalignment='top', x=0.51, y=0.965)

plt.savefig('plots/convergance_forest', bbox_inches="tight")
plt.show()
```
